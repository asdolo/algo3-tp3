```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
savingsTiemposA = pd.read_csv('1-savings/A/tiempo.csv')
savingsOptimosA = pd.read_csv('1-savings/A/optimo.csv')
savings = savingsTiemposA.merge(savingsOptimosA, on='n')
golosaTiemposA = pd.read_csv('2-golosa/A/tiempo.csv')
golosaOptimosA = pd.read_csv('2-golosa/A/optimo.csv')
golosa = golosaTiemposA.merge(golosaOptimosA, on='n')
```

```python
#TEST Complejidad savings (Uso instancias de clase A)
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

savings['SAV/N^3'] = savings['tiempo'] / savings['n']**3;
ax1 = sns.lineplot(x="n", y="SAV/N^3", data=savings.reset_index(), label="Saving", ax=ax);

#Reverse legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(reversed(handles), reversed(labels), title='LÃ­neas', loc='upper right')

plt.ylim(-0.02,0.02);

plt.ylabel(r"F(G,N) = $\frac{tiempo(G,N)}{N^3}$", fontsize=16);
plt.xlabel("N", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
#plt.savefig("FByBA_Complejidad.png")
```

```python
#TEST Resultado vs optimo (Uso instancias de clase A)
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

ax1 = sns.lineplot(x="n", y="resultado", data=savings.reset_index(), label="Saving", ax=ax);
ax2 = sns.lineplot(x="n", y="optimo", data=savings.reset_index(), label="Optimo", ax=ax);

#Reverse legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(reversed(handles), reversed(labels), title='LÃ­neas', loc='upper right')

#plt.ylim(-0.01,0.01);

plt.ylabel(r"Distancia recorrida", fontsize=16);
plt.xlabel("N", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
#plt.savefig("FByBA_Complejidad.png")
```

```python
#TEST Bar plot Diferencias entre Resultado vs optimo (Uso instancias de clase A)
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
savings['distanciaOptRes'] = savings['resultado'] / savings['optimo']
ax1 = sns.barplot(x="n", y="distanciaOptRes", data=savings,ax=ax)
#ax1 = sns.lineplot(x="n", y="resultado", data=savings.reset_index(), label="Saving", ax=ax);


#Reverse legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(reversed(handles), reversed(labels), title='LÃ­neas', loc='upper right')

#plt.ylim(-0.01,0.01);

plt.ylabel(r"distancia(Saving) - distancia(optimo)", fontsize=16);
plt.xlabel("N", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
#plt.savefig("FByBA_Complejidad.png")
```
